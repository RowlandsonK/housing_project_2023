# Cleaning

```python
# Identify and display duplicate rows based on the 'id' column.

kc_house_data[kc_house_data.duplicated(subset='id', keep=False)]
```

```python
# Remove duplicate rows, keeping only the last occurrence based on the 'id' column. keep last date sold

kc_house_data.drop_duplicates(subset='id', keep='last', inplace=True)
```

#### Price and ID Columns

```python
# Reorder columns so that the target variable is the first column
# Exclude 'id' column

column_headers = list(kc_house_data.columns)
column_headers = [column_headers[2]]+[column_headers[1]]+column_headers[3:]
kc_house_data = kc_house_data[column_headers]
```

```python
# Change the data type of the 'price' column to integer.

kc_house_data = kc_house_data.astype({'price':'int'})
```

#### Date Column

```python
# Split 'date' column into 'day', 'month', and 'year' columns

kc_house_data.date = pd.to_datetime(kc_house_data.date)
kc_house_data['month'] = kc_house_data.date.dt.month

# Reorder columns again
# Exclude 'date' column

column_headers = list(kc_house_data.columns)
column_headers = [column_headers[0]]+[column_headers[20]]+column_headers[2:20]
kc_house_data = kc_house_data[column_headers]
```

#### Bedrooms Column

```python
# Create a boxplot to visualize the distribution of the 'bedrooms' column.

sns.boxplot(x = kc_house_data.bedrooms);
```

```python
# Remove an extreme outlier in the 'bedrooms' column where the value is 33.

kc_house_data.drop(kc_house_data[kc_house_data.bedrooms == 33].index, inplace=True)
```

#### Bathrooms Column

```python
# Create a histogram to visualize the distribution of the 'bathrooms' column.

sns.histplot(kc_house_data.bathrooms)
```

```python
# Remove rows where the 'bathrooms' column has less than 1 bathroom.

kc_house_data.drop(kc_house_data[kc_house_data.bathrooms < 1].index, inplace=True)
```

#### Waterfront Column

```python
# Display the counts and normalized counts (percentages) of unique values in the 'waterfront' column.

kc_house_data.waterfront.value_counts()
```

```python
# Display the updated counts and normalized counts of unique values in the 'waterfront' column after filling NaN values.

kc_house_data.waterfront.value_counts(normalize=True)
```

```python
probabilities = list(kc_house_data.waterfront.value_counts(normalize=True).values)
unique_values = kc_house_data.waterfront.dropna().unique()
np.random.seed(0)

def replace_nan_with_random(x):
    if pd.isna(x):
        return np.random.choice(unique_values, p=probabilities)
    else:
        return x

kc_house_data.waterfront = kc_house_data.waterfront.apply(replace_nan_with_random)
```

#### View Column

```python
# Display the counts of unique values in the 'view' column.

kc_house_data.view.value_counts()
```

```python
# Compute probabilities for each value in the 'view' column

view_probabilities = kc_house_data['view'].value_counts(normalize=True).to_dict()

# Function to replace NaN with a random choice based on probabilities

def replace_nan_with_random(x):
    if pd.isna(x):
        return np.random.choice(list(view_probabilities.keys()), p=list(view_probabilities.values()))
    else:
        return x

# Apply the function to the 'view' column

kc_house_data['view'] = kc_house_data['view'].apply(replace_nan_with_random)
```

```python
# Define a mapping dictionary for known values

mapping = {'GOOD': 1, 'NONE': 2}

# Function to map values using the dictionary or set to -1 if not found

def map_with_default(x):
    return mapping.get(x, -1)

# Replace non-integer values with their corresponding codes or -1 for unknown values

kc_house_data['view'] = kc_house_data['view'].apply(map_with_default)

# Convert the column to an integer type

kc_house_data = kc_house_data.astype({'view': 'int'})
```
